```python
#create a subset
df = df_ords_prods_merged[:1000000]
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#apply the function 
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#check the frequencies
df['price_range'].value_counts(dropna = False)
```

```python
#Check what is the most expensive product
df['prices'].max()
```

```python
#apply the function using loc
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#Create a new column, Busiest days
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Busiest day")
  elif value == 4 or value == 3:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Combine with df_ords_prods_merged
df_ords_prods_merged['busiest_days'] = result
```

### We can see that the values are different from when we only had one busiest and slowest day. Previously, more days were labeled as "regular busy", whereas now, the days that were previously "regular busy", are either busiest or least busy. 

```python
#check the frequency of order_hour_of_day to find busiest/least busy hour
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

### We can see that there are 24 hours, so we could divide them into 3 sections, most, average and fewest orders. We can assign the top 8 values to the most orders, the bottom 8 to the fewest, and the leftovers to the average




```python
#Use for-loop statement 
hour_of_day = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value == 10 or value ==11 or value ==14 or value ==15 or value ==13 or value ==12 or value ==16 or value ==9:
    hour_of_day.append("Most orders")
  elif value ==3 or value ==4 or value ==2 or value ==5 or value ==1 or value ==0 or value ==6 or value ==23:
    hour_of_day.append("Fewest orders")
  else:
    hour_of_day.append("Average orders")
```

```python
#Create a new column
df_ords_prods_merged['busiest_time_of_day'] = hour_of_day
```
